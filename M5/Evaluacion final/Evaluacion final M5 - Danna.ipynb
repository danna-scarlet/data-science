Los  resultados  del  rendimiento  académico  se  miden  mediante  un  examen  estándar,  donde  las calificaciones oscilan entre 0 y 100 puntos. Los datos obtenidos son los siguientes:

Grupo B (Control): 70, 72, 75, 78, 80, 68, 74, 76, 79, 77, 73, 71, 75, 78, 80

```python
#2.Cálculo de Estadísticas Descriptivas (3puntos)  
#•Calcula la media y la desviación estándar para ambos grupos.  

import pandas as pd
data = {'Grupo_A': [85, 90, 78, 88, 92, 80, 86, 89, 84, 87, 91, 82, 83, 85, 88],
        'Grupo_B': [70, 72, 75, 78, 80, 68, 74, 76, 79, 77, 73, 71, 75, 78, 80]}

df = pd.DataFrame(data)

# Calcular la media y desviación estándar de 'Grupo_A'
media_grupo_a = df['Grupo_A'].mean()
desviacion_grupo_a = df['Grupo_A'].std()
print(f"La media de Grupo A (tutoría) es: {media_grupo_a}")
print(f"La desviación estándar de Grupo A (tutoría) es: {desviacion_grupo_a}")

# Calcular la media y desviación estándar de 'Grupo_B'
media_grupo_b = df['Grupo_B'].mean()
desviacion_grupo_b = df['Grupo_B'].std()
print(f"La media de Grupo B (Control) es: {media_grupo_b}")
print(f"La desviación estándar de Grupo B (Control) es: {desviacion_grupo_b}")
```

```python
#•Representa los datos gráficamente (usando histogramas o diagramas de caja). 
import matplotlib.pyplot as plt

# Histograma para Grupo A
plt.figure(figsize=(12, 6))
plt.hist(df['Grupo_A'], bins=15, color='blue', alpha=0.7, rwidth=0.8)
plt.title('Histograma de Grupo A (tutoría): Cantidad de alumnos v/s puntuaciones')
plt.xlabel('Puntuaciones')
plt.ylabel('Cantidad de alumnos')

# Histograma para Grupo B
plt.figure(figsize=(12, 6))
plt.hist(df['Grupo_B'], bins=15, color='green', alpha=0.7, rwidth=0.8)
plt.title('Histograma de Grupo B (Control): Cantidad de alumnos v/s puntuaciones')
plt.xlabel('Puntuaciones')
plt.ylabel('Cantidad de alumnos')

plt.tight_layout()
plt.show()
```

```python
# Diagrama de caja para Grupo A
plt.boxplot(df['Grupo_A'], vert=False)
plt.title('Diagrama de Caja de Grupo A (tutoría)')
plt.xlabel('Puntuaciones')
plt.tight_layout()
plt.show()

# Diagrama de caja para Grupo B
plt.boxplot(df['Grupo_B'], vert=False)
plt.title('Diagrama de Caja de Grupo B (Control)')
plt.xlabel('Puntuaciones')
plt.tight_layout()
plt.show()
```

```python
#3.Prueba de Hipótesis (4 puntos)  
#•Plantea la hipótesis nula(H0) y alternativa (H1). 
# H0: No hay diferencia en el rendimiento académico entre los dos grupos. H1: El grupo con tutoría tiene un mejor rendimiento académico. 

#H0: puntaje promedio de grupo a (con tutoria) es igual a 75.06666666666666 (grupo b de control)
#H1: puntaje promedio de grupo a (con tutoria) es mayor a 75.06666666666666 (grupo b de control)

#•Realiza  una  prueba  t  para  comparar  las  medias  de  ambos  grupos.  Usa  un  nivel  de significancia (α=0.05). 
#•Interpreta el valor-p y decide si rechazas o no la hipótesis nula.  

alpha = 0.05

import scipy.stats as stats

media_hipotetica = media_grupo_b
muestra = df['Grupo_A']
statistic, p_value = stats.ttest_1samp(muestra, popmean=media_hipotetica)
print(f'Estadístico t: {statistic:.3f}')
print(f'Valor p: {p_value:.4f}')

if p_value < alpha:
    print('Rechazamos H0: Hay evidencia de que el puntaje promedio de grupo a (con tutoria) es mayor a 75.06666666666666 (grupo b de control)')
else:
    print('No rechazamos H0: No hay evidencia suficiente para decir que el puntaje promedio de grupo a (con tutoria) es mayor a 75.06666666666666 (grupo b de control)')
```

```python
#4.Intervalo de Confianza (2 puntos)  
#•Calcula un intervalo de confianza del 95% para la diferencia de medias entre los dos grupos.  
#•Interpreta el resultado.  

import numpy as np
from scipy.stats import t

#Parametros
X_bar = media_grupo_a
n = len(df['Grupo_A'])
s = desviacion_grupo_a
confianza = 0.95

#Calcular grados de libertad
grados_libertad = n - 1

#Calcular valor critico t_alpha/2
t_critico = t.ppf(1 - (1 - confianza) / 2, df=grados_libertad)

#calcular margen de error
margen_error = t_critico * (s / np.sqrt(n))

#calcular limites del intervalo de confianza
limite_inferior = X_bar - margen_error
limite_superior = X_bar + margen_error

#resultados
print(f'Valor critico t: {t_critico:.3f}')
print(f'Margen de error: {margen_error:.2f}')
print(f'Intervalo de confianza del 95% para la media del grupo A: ({limite_inferior:.2f}, {limite_superior:.2f})')
```
